```python
_ = """
MARS - Multivariable Adaptive Regression Splines

1) Forward:
    x1, ... xp -> x'1, ... x'l (l > p)
2) Backward
    Select a subset of x'1, ... x'l
3) Model

Forward step:
    (x - t) = max(x-t, 0)  # RELU
    (t - x) = max(t-x, 0)  # Odbicie lustrzane RELU

    x'm+1 = x'l * (xv - t)
    x'm+2 = x'l * (t - xv)
"""
```

```python
import numpy as np
```

```python
def get_dataset_1(n: int = 1000, p: int = 50) -> np.array:
    x1 = np.array([
        np.random.uniform(0, 4)
        for _ in range(n)
        ]).reshape(-1, 1)

    x2_p = np.array([
        [np.random.normal(0, 1) for _ in range(p)]
        for _ in range(n)
    ])

    y = np.array([
        x_i**(1/2) + np.random.normal(0, 0.1)
        for x_i in x1
    ]).reshape(-1, 1)

    return np.hstack([x1, x2_p, y])


def get_dataset_2(n: int = 1000, p: int = 50) -> np.array:
    x1 = np.array([
        np.random.uniform(0, 4)
        for _ in range(n)
        ]).reshape(-1, 1)

    x2_p = np.array([
        [np.random.normal(0, 1) for _ in range(p)]
        for _ in range(n)
    ])

    y = np.array([
        x_i**2 + np.random.normal(0, 0.1)
        for x_i in x1
    ]).reshape(-1, 1)

    return np.hstack([x1, x2_p, y])


def get_dataset_3(n: int = 1000, p: int = 50) -> np.array:
    x = np.array([
        [np.random.normal(0, 1) for _ in range(p)]
        for _ in range(n)
    ])

    y = np.array([
        max((x_i[0],  0)) + max((x_i[0] - 1, 0)) + np.random.normal(0, 0.1)
        for x_i in x
    ]).reshape(-1, 1)

    return np.hstack([x, y])


def get_dataset_4(n: int = 1000, p: int = 50) -> np.array:
    x1 = np.array([
        np.random.uniform(0, 4)
        for _ in range(n)
        ]).reshape(-1, 1)

    x2_p = np.array([
        [np.random.normal(0, 1) for _ in range(p)]
        for _ in range(n)
    ])

    y = np.array([
        np.sin(x_i) + np.random.normal(0, 0.1)
        for x_i in x1
    ]).reshape(-1, 1)

    return np.hstack([x1, x2_p, y])


# def get_dataset_5(n: int = 1000, p: int = 50) -> np.array:
#     x = np.array([
#         [np.random.normal(0, 1) for _ in range(p)]
#         for _ in range(n)
#     ])

#     y = np.array([
#         x_i
#         for x_i in x
#     ]).reshape(-1, 1)

#     return np.hstack([x, y])
```
